### Imports

```python
from bs4 import BeautifulSoup
import requests
```

### HTTP Request

```python
# Store our website in a variable
```

```python
website = 'http://automationpractice.com/index.php?id_category=3&controller=category'
```

```python
response =requests.get(website)
```

```python
# Check the status code
response.status_code
```

### Soup Object

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
print(soup)
```

### Important Functions

#### find() 

returns first element

```python
# example 1 - find by id 
```

```python
contact_us=soup.find(id='contact-link').get_text().replace('\n','')
contact_us
```

```python
# example 2 - find by id
```

```python
contact_us_2 = soup.find(id='contact-link').find('a').get_text()
contact_us_2
```

```python
# example 3 - find by class
result = soup.find(class_="ajax_block_product").find(class_="content_price")
result
```

#### find_all() & findAll()

returns a list

```python
result_ = soup.find_all(class_="ajax_block_product")
result_
```

```python
len(result_)
```

```python
result_x = soup.findAll(class_="ajax_block_product")
result_x
```

```python
len(result_x)
```

#### select_one()

select by css selectors - returns first element - equivalent to "find()"

```python
soup.select_one(".shop-phone").get_text().strip().replace("Call us now: ",'')  # whenever we want to target class use"."
```

```python
soup.select_one('#contact-link').get_text().strip()  # whenever we want to target id use"#".
```

#### select()

select by css selectors - returns a list - equivalent to "find_all()/findAll()"

```python
soup.select(".ajax_block_product")
```

```python
len(soup.select(".ajax_block_product"))
```

#### get value of attributes vs. get_text

```python
# find by attributes
soup.find(itemprop='name').find("a").get("title")
```

```python
soup.find(itemprop='name').find("a").get_text().strip()
```

### Siblings and Parents

```python
# go to product #1 and find the next sibling
soup.select_one(".ajax_block_product").find_next_sibling().select_one('.product-name').get_text().strip()
```

```python
len(soup.select_one(".ajax_block_product").find_next_siblings())
```

```python
# go to product #7 (last product) and find the previous sibling
soup.select(".ajax_block_product")[6].find_previous_sibling().select_one(".product-name").get_text().strip()
```

```python
# go to last product and find the parent element
soup.select(".ajax_block_product")[6].find_parent()
```

### Extract Links

#### Single Link

```python
soup.find(id="contact-link").find('a')['href']
```

#### multipe links

```python
link_variable = soup.find_all(itemprop='name')
link_variable
```

```python
links=[]

for i in link_variable:
    links.append(i.find('a')['href'])
```

```python
links
```
